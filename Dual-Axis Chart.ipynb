# Timeseries Chart | Multiple axis

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# create a dataframe with the provided data
data = {'Date': ['2/1/2023', '2/1/2023', '2/1/2023', '2/2/2023', '2/2/2023', '2/2/2023', '2/3/2023', '2/3/2023', '2/3/2023', '2/4/2023', '2/4/2023', '2/4/2023', '2/5/2023', '2/5/2023', '2/5/2023', '2/6/2023', '2/6/2023', '2/6/2023'],
        'Category': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green'],
        'Count': [100, 50, 200, 75, 75, 200, 150, 100, 150, 200, 75, 175, 250, 125, 150, 250, 175, 100]}

df = pd.DataFrame(data)
```

```python
# Set plot style and size
mpl.rcParams['figure.figsize'] = [16.0, 10.0]
mpl.style.use('ggplot')

# convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# group the dataframe by 'Category' and 'Date' and sum the counts
df_grouped = df.groupby(['Category', 'Date']).sum()

# pivot the dataframe to create a separate column for each category
df_pivoted = df_grouped.pivot_table(index='Date', columns='Category', values='Count')

# set line colors
colors = ['r', 'b', 'g']

# create a line plot of the data
fig, ax = plt.subplots()
ax.set_facecolor('w')
df_pivoted.plot.line(color=colors, ax=ax)

# add labels and titles
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Category Count Over Time')

# display the plot
plt.show()
```
